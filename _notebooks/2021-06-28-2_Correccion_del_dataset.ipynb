## IMPUTACIÓN DE VALORES FALTANTES

Del analisis hecho en la parte I del proyecto, no se observaron claros valores erroneos o *outliers*. Luego, la parte de limpieza del *dataset* consiste principalmente en reemplazar los valores faltantes (NaNs).

**NOTA:** A lo largo de este *notebook* muy frequentemente utilizaré el valor $-123456789$ como sustituto de los NaNs.  Esto es una estrategia que utilizo para poder tratar los datos. Muchas veces realizo una operación sobre los valores solo si se cumple una condicion logica:

* $if\,(condicion)\;\;\;\;\;\;\;\;\;=>\;\;Valor_A$
* $if\,(not\,(condicion))\;\;=>\;\;Valor_B$

Codificando el resultado de la operación lógica como $0$ o $1$, podemos escribirlo como:

* $(if\,(condicion))\;x\;Valor_A\;+\;if\,(not\,(condicion))\;x\,Valor_B$

Esto es igual a $Valor_A$ si se cumple la condición o $Valor_B$ en caso contrario. 

Trabajar de esta forma permite operar sobre el dataset de manera más rápida. El problema es que NaN x 0 = NaN. Es por ello que uso un "NaN" numérico ($-123456789$) que sé que no aparece en el *dataset*

```python
#hide
_nan=-123456789
```

### 0) Generacion de un clasificador de competiciones en: liga y copa

Muchas veces será de interes saber si un "league_id" corresponde a una competición de copa o liga. Para ello, utilizando los partidos para los cuales tenemos la informacion sobre el tipo de competición ("is_cup") , generamos una suerte de identificador a través de un diccionario, donde el *key* es el "id" y el valor $1$ o $0$ si es una copa o liga, respectivamente.

Patra ello utilizaremos tanto los datos históricos como los de los partidos a modelar.

```python
#collapse-show
ID=FMP['league_id'].to_numpy()
TIPO=FMP['is_cup'].to_numpy()*1.0
ID_fil=FMP['league_id'].isna().to_numpy()
TIPO_fil=FMP['is_cup'].isna().to_numpy()


for t in ['home','away']:
    for i in range(1,11):
        l1=t+'_team_history_league_id_{}'.format(i)
        l2=t+'_team_history_is_cup_{}'.format(i)
        
        ID=np.concatenate((ID,FMP[l1].to_numpy()),axis=0)
        TIPO=np.concatenate((TIPO,FMP[l2].to_numpy()),axis=0)
        ID_fil=np.concatenate((ID_fil,FMP[l1].isna().to_numpy()),axis=0)
        TIPO_fil=np.concatenate((TIPO_fil,FMP[l2].isna().to_numpy()),axis=0)

ID=ID[~TIPO_fil]
ID_fil=ID_fil[~TIPO_fil]
TIPO=TIPO[~TIPO_fil]

ID=ID[~ID_fil]
TIPO=TIPO[~ID_fil]


COMPETICIONES=defaultdict(lambda: _nan,zip(ID,TIPO)) 
```

### 3) Identificador de competición

Para reemplazar los valores faltantes en la columna del identificador de la competición ("league_id"), lo que hicimos fue completarlo considerando el valor más probable. 

Para hacer ello, primero hay que recordar que que para cada equipo tenemos $10$ partidos de historial. Luego podemos buscar cual es la competición con más ocurencias en esa muestra de diez partidos y reemplazar los valores faltantes con ese "id".  Como precaución, se require tener una muestra igual o mayor a $4$ y tener un único valor con más ocurrencias.

En los casos de los equipos para los cuales no se cumplan las condiciones, se completarán los "id" con $-1$.

```python
#collapse-show
for t in ['home','away']:
    L=[]
    for i in range(1,11):
        l=t+'_team_history_league_id_{}'.format(i)
        L.append(l)

    league_id=FMP[L]
    
    FMP[t+'_league_id_nans']=np.isnan(league_id).sum(axis=1)

    FMP[t+'_most_common']=league_id.mode(axis=1)[0]
    
    FMP[t+'_most_common_n']=np.isnan(league_id.mode(axis=1)).sum(axis=1)
    
    FMP[t+'_most_common'].fillna(-1, inplace = True) 
    
    FMP[t+'_most_common']=FMP[t+'_most_common']*(FMP[t+'_most_common_n']==1).astype('int') + (-1)*(FMP[t+'_most_common_n']>1).astype('int')
```

```python
#collapse-show
maximo_de_nans=6
for t in ['home','away']:
    nans=t+'_league_id_nans'
    most_common=t+'_most_common'
    for i in range(1,11):
        c=t+'_team_history_league_id_{}'.format(i)    
        
        FMP[c].fillna(_nan, inplace = True)
            
        FMP[c]=((FMP[c]==_nan) & (FMP[nans]<=maximo_de_nans)).astype('int')*FMP[most_common]+( (FMP[c]!=_nan) | (FMP[nans]>maximo_de_nans)).astype('int')*FMP[c]
        
        FMP[c]=FMP[c].replace(to_replace=_nan, value=np.nan)
```

```python
#collapse-show
#rellenamos los faltante con -1 y dejamos todo como enteros:
for t in ['home','away']:
    for i in range(1,11):
        c=t+'_team_history_league_id_{}'.format(i)
        FMP[c].fillna(-1, inplace = True)
        FMP[c]=FMP[c].astype('int')
```

En este caso, no hay identificadores faltantes para los partidos a modelar. Sin embargo, he pensado como se haría en dicho caso. La mejor opción sería fijarse si tienen alguna competición en comun. Para ello habría que tener una lista de las competiciones para el equipo local y visitante, lo que sería un tanto complejo. Como opción B, se podría al menos comparar la competición con más correncias de cada uno, y, si coincide, tomar ese valor. Esto es lo que muestro abajo.

```python
#collapse-show
c='league_id'
most_common_home='home_most_common'
most_common_away='away_most_common'

FMP[c].fillna(_nan, inplace = True)
FMP[c]=( (FMP[c]==_nan) & (FMP[most_common_home]==FMP[most_common_away])).astype('int')*FMP[most_common_home]+(  (FMP[c]!=_nan) | (FMP[most_common_home]!=FMP[most_common_away])).astype('int')*FMP[c]
FMP[c]=FMP[c].replace(to_replace=_nan, value=np.nan)

FMP.league_id.fillna(-1, inplace = True)
```

### 6) Identificador del entrenador

Como marco general, se debe notar que esta variable la utilizaremos solamente para registrar los partidos donde se produzca un cambio de entrenador. Esto se debe a que dicho evento podría generar un efecto anímico en los equipos que repercuta en sus resultados.

Luego, comenzando por el partido $i=9$ hasta llegar al partido a modelar, la estrategia de imputación utilizada fue la siguiente:

i) Para los casos en los que sabía el id del siguente partido y del anterior:
* Si son iguales: Reemplazaré por ese id
* Si son distintos: reemplazaré por el anterior (no puedo saber si cambio en ese partido o el siguiente)

ii) Para los casos donde tenía solo el valor anterior y no el siguiente, utilizaré el anterior (no hay razón para presuponer un cambio de tecnico)

iii) Finalmente, en los casos donde no tenía ningún valor contiguo, reemplacé por $-1$ (dado que los id son enteros positivos).

En resumen, reemplacé por el valor anterior, siempre que exista. Caso contrario, imputé con un $-1$.

Finalmente, para el partido $i=10$, en función de la estrategia adoptada, la unica opción fue imputar con $-1$.

```python
#collapse-show
for t in ['home','away']:
    for i in range(9,0,-1):
        c=t+'_team_history_coach_{}'.format(i)
        a=t+'_team_history_coach_{}'.format(i+1)
        
        FMP[c].fillna(_nan, inplace = True)
        FMP[a].fillna(_nan, inplace = True)
        FMP[c]=( (FMP[c]==_nan) & (FMP[a]!=_nan) ).astype('int')*FMP[a]+( (FMP[c]!=_nan) | (FMP[a]==_nan) ).astype('int')*FMP[c]
        FMP[c]=FMP[c].replace(to_replace=_nan, value=np.nan)
        FMP[a]=FMP[a].replace(to_replace=_nan, value=np.nan)
            
for t in ['home','away']:
    c=t+'_team_coach_id'
    a=t+'_team_history_coach_{}'.format(1)
    
    FMP[c].fillna(_nan, inplace = True)
    FMP[a].fillna(_nan, inplace = True)
    FMP[c]=( (FMP[c]==_nan) & (FMP[a]!=_nan) ).astype('int')*FMP[a]+( (FMP[c]!=_nan) | (FMP[a]==_nan) ).astype('int')*FMP[c]
    FMP[c]=FMP[c].replace(to_replace=_nan, value=np.nan)
    FMP[a]=FMP[a].replace(to_replace=_nan, value=np.nan)
```

Dado que en este caso no podía imputar simplemente mirando a la columna y sacando una mediana o media, ya que el *dataset* contiene grupos de diez partidos de 2019 a 2021, tomé una estrategia diferente.

Empecé por calcular un tiempo "medio" de los tiempos entre los partidos. Para ello, primero obtube las medianas de los tiempo entre los partidos, comenzando por $i_{partido\,a\,modelar} - i_1$ hasta $i_9 - i_{10}$. Luego, tomé una media de esos valores pesada por la cantidad de datos utilizados en cada caso. 

```python
#collapse-show
M=[]
N=[]
for t in ['home','away']:
    for i in range(1,10):
        c=t+'_team_history_match_date_{}'.format(i)
        a=t+'_team_history_match_date_{}'.format(i+1)
        
        D=FMP[[c,a]]
        D.dropna(inplace=True)
       
        D=D[c]-D[a]
        
        m=D.median()#.days 
        M.append(m)
        n=len(D)
        N.append(n)
for t in ['home','away']:
    c='match_date'
    a=t+'_team_history_match_date_{}'.format(1)

    D=FMP[[c,a]]
    D.dropna(inplace=True)

    D=D[c]-D[a]

    m=D.median()#.days 
    M.append(m)
    n=len(D)
    N.append(n)

M=np.array(M)
N=np.array(N)#funciona como una especia de peso para mada mediana calculada
Tiempo_entre_partidos=np.sum(M*(N/np.sum(N)))
print('Tiempo entre partidos: ',Tiempo_entre_partidos)
```

De mi analisis de los valores faltantes, sabía que teníamos todas las fechas correspondientes a los partidos a modelar. Luego, comezando por $i=1$ hasta $i=10$, proseguí a imputar todos los valores faltantes, por la fecha del partido siguiente menos el valor calculado. Como se puede ver en la tabla al final, esta estrategia me sirvió para corregir todos los valores necesarios.

```python
#collapse-show
T=[]
I=[]
A=[]
D=[]
for t in ['home','away']:
    p='match_date'
    c=t+'_team_history_match_date_{}'.format(1)
    
    antes=FMP[c].isna().sum()
    FMP[c] = FMP.apply(lambda row: row[p]-Tiempo_entre_partidos if (pd.isna(row[c]) & (~(pd.isna(row[p])))) else row[c],axis=1)
    despues=FMP[c].isna().sum()
    
    T.append(t)
    I.append(1)
    A.append(antes)
    D.append(despues)

for t in ['home','away']:
    for i in range(2,11):
        p=t+'_team_history_match_date_{}'.format(i-1)
        c=t+'_team_history_match_date_{}'.format(i)

        antes=FMP[c].isna().sum()
        FMP[c] = FMP.apply(lambda row: row[p]-Tiempo_entre_partidos if (pd.isna(row[c]) & (~(pd.isna(row[p])))) else row[c],axis=1)
        despues=FMP[c].isna().sum()

        T.append(t)
        I.append(i)
        A.append(antes)
        D.append(despues)

print('Datos erroneos')
display(pd.DataFrame({'Team':T,'Partido':I,'Antes':A,'Despues':D,'Resueltos':np.array(A)-np.array(D)}))        
```

### 9) Goles

En este caso tenemos los goles que marco cada equipo (*away* y *home*) y sus oponentes, para cada uno de los $10$ partidos. 

Para corregir estos valores, teniendo en cuenta el análisis hecho en el bloque exploratorio 1), procedí de la siguiente manera:

* Para el "*team (away/home)*", si tenía 6 o más partidos con datos de los goles marcados, sortee un valor con distribución de Poison con $\lambda$ igual a su media de goles en esos partidos.

* Caso contrario y para los oponentes, haciendo uso de lo que vimos al explorar las variables, generamos valores de los goles considerando una distribución de poisson con:
    * $\lambda=1.52$ para los locales
    * $\lambda=1.24$ para los visitantes

```python
#collapse-show
for t in ['home','away']:
    L=[]
    for i in range(1,11):
        l=t+'_team_history_goal_{}'.format(i)
        L.append(l)
    m=np.nanmean(FMP[L],axis=1)
    n=np.isnan(FMP[L]).sum(axis=1)
    
    c=t+'_team_mean_goal'
    FMP[c]=m
    FMP[c].fillna(_nan, inplace = True)
    
    c=t+'_nans_for_goal'
    FMP[c]=n
```

### 10) *Rating* de los Equipos

El panorama aquí es bastantesimilar a lo que teníamos para los goles. Nuevamente, tenemos el *rating* de cada equipo (*away* y *home*) y sus oponentes, para cada uno de los $10$ partidos.

Para corregir estos valores procedí de la siguiente manera:
* Para el "team (away/home)" utilicé la media, siempre que tuviéramos más de 5 partidos
* Caso contrario y para los oponentes, generé un rating basandome en la distribución log-norm encontrada en el análisis exploratorio, distinguiendo nuevamente entre locales y visitantes.

```python
#collapse-show
for t in ['home','away']:
    L=[]
    for i in range(1,11):
        l=t+'_team_history_rating_{}'.format(i)
        L.append(l)
    m=np.nanmedian(FMP[L],axis=1)
    n=np.isnan(FMP[L]).sum(axis=1)
    
    c=t+'_team_median_rating'
    FMP[c]=m
    FMP[c].fillna(_nan, inplace = True)
    
    c=t+'_nans_for_rating'
    FMP[c]=n
```

```python
#collapse-show
maximo_de_nans=5
for t in ['home','away']:
    r=t+'_team_median_rating'
    n=t+'_nans_for_rating'
    for i in range(1,11):
        l=t+'_team_history_rating_{}'.format(i)
        
        FMP[l].fillna(_nan, inplace = True)
        
        FMP[l]=(  (FMP[n]<maximo_de_nans) & (FMP[l]==_nan) ).astype('int')*FMP[r] + (  (FMP[n]>=maximo_de_nans) | (FMP[l]!=_nan) ).astype('int')*FMP[l]
        
        FMP[l]=FMP[l].replace(to_replace=_nan, value=np.nan)
```

```python
#collapse-show
m_home=7.14
std_home=2.68
m_l_home=np.math.log(m_home**2/(np.math.sqrt(m_home**2+std_home*2)))
std_l_home=np.math.sqrt(np.math.log(1+((std_home**2)/(m_home**2))))

m_visitor=5.87
std_visitor=2.41
m_l_visitor=np.math.log(m_visitor**2/(np.math.sqrt(m_visitor**2+std_visitor*2)))
std_l_visitor=np.math.sqrt(np.math.log(1+((std_visitor**2)/(m_visitor**2))))

lengh=len(FMP)
```

```python
#collapse-show
for t in ['home','away']:
    for i in range(1,11):
        c_team=t+'_team_history_rating_{}'.format(i)
        c_opponent=t+'_team_history_opponent_rating_{}'.format(i)
        l=t+'_team_history_is_play_home_{}'.format(i)
        
        
        Rdm_rating_local = pd.Series(np.random.lognormal(m_l_home, std_l_home, lengh),index=FMP.index)
        Rdm_rating_visitor = pd.Series(np.random.lognormal(m_l_visitor, std_l_visitor, lengh),index=FMP.index)
        
        
        FMP[c_team].fillna(_nan, inplace = True)
        FMP[c_team]=( (FMP[c_team]==_nan) & (FMP[l]==1) ).astype('int')*Rdm_rating_local+( (FMP[c_team]==_nan) & (FMP[l]==0) ).astype('int')*Rdm_rating_visitor+(FMP[c_team]!=_nan).astype('int')*FMP[c_team]
        FMP[c_team]=FMP[c_team].replace(to_replace=_nan, value=np.nan)
        
        FMP[c_opponent].fillna(_nan, inplace = True)
        FMP[c_opponent]=( (FMP[c_opponent]==_nan) & (FMP[l]==0) ).astype('int')*Rdm_rating_local+( (FMP[c_opponent]==_nan) & (FMP[l]==1) ).astype('int')*Rdm_rating_visitor+(FMP[c_opponent]!=_nan).astype('int')*FMP[c_opponent]
        FMP[c_opponent]=FMP[c_opponent].replace(to_replace=_nan, value=np.nan)
```

```python
#collapse-show
r='home_team_median_rating'
FMP[r]=FMP[r].replace(to_replace=_nan, value=np.nan)
Rdm_rating_local = pd.Series(np.random.lognormal(m_l_home, std_l_home, lengh),index=FMP.index)
FMP[r].fillna(Rdm_rating_local, inplace = True)

r='away_team_median_rating'
FMP[r]=FMP[r].replace(to_replace=_nan, value=np.nan)
Rdm_rating_visitor = pd.Series(np.random.lognormal(m_l_visitor, std_l_visitor, lengh),index=FMP.index)
FMP[r].fillna(Rdm_rating_visitor, inplace = True)

for t in ['home','away']: 
    n=t+'_nans_for_rating'
    FMP=FMP.drop(columns=[n])
```
